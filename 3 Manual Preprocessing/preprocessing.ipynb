```python
# Replace missing values cat -mode, con- mean/median1
def replacer(df):
    cat = list(df.columns[df.dtypes == "object"])

    for col in df.columns:
        if col in cat:
            mode = df[col].mode()[0]
            df[col] = df[col].fillna(mode)
        else:
            mean = df[col].mean()
            df[col] = df[col].fillna(mean)
```

## Apply one hot encoding on X_cat

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(
    handle_unknown="ignore", sparse_output=False
).set_output(transform="pandas")

X_cat_pre = ohe.fit_transform(X_cat)
```

```python
X_cat_pre.head()
```

### Scaling on numerical features
standard scalar is preffered technique

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().set_output(transform="pandas")
scaler.fit_transform(X_con)
X_con_pre = scaler.fit_transform(X_con)
X_con_pre
```

### Combine the cat and con var

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre
```

### Step 5 - Seperate X_pre and Y_pre

```python
from sklearn.model_selection import train_test_split

xtrain , xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

## Step 6 - Build the model


```python
from sklearn. linear_model import LinearRegression
model = LinearRegression ()
model. fit(xtrain, ytrain)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model, x, y):
    # Predict results for x
    ypred = model.predict(x)
    
    # Calculate the metrics
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)

    # Print above results
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
print("Training Results : ")
evaluate_model(model, xtrain, ytrain)
```

```python
print("Testing Results")
evaluate_model(model, xtest, ytest)
```

### R2 score is greater than 80% hence model is good model
